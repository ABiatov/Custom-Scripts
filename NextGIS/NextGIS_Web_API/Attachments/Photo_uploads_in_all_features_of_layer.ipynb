# Import libraries

```python
import requests
import json
import os
```

# NGW connection

```python
# Web GIS connection parameters
ngw_url = 'https://<WEB-GIS URL>.nextgis.com'
ngw_api_url = ngw_url + '/api'
ngw_auth = ('USER', 'UOUR PASSWORD')

curent_resource = 170 #layer id

# Layer URL
feature_dst = ngw_api_url+'/resource/' + str(curent_resource) + '/feature/'   
# feature_dst
```

```python
# Data folder
photo_directory = 'D:/MAPS/WD/rivers/photo'
```

```python
# get all features
my_fs = requests.get(feature_dst)
my_fs_json = my_fs.json()
my_fs_json[8]
```

```python
len(my_fs_json)
```

```python
# Filtering features, if necessary

my_fs_filtred_json = []

for elem in my_fs_json:
#    if elem['fields']['file_list'] != None:   # Feature filtering rule, if necessary
    if (elem['fields']['file_list'] != None and elem['id'] > 50):  # Feature filtering rule, if necessary
#        print(elem['id'])
        my_fs_filtred_json.append(elem)

# my_fs_filtred_json[8]
```

```python
len(my_fs_filtred_json)
```

```python
for elem in my_fs_filtred_json:
    feature_id = elem['id']
    r = requests.get(feature_dst + str(feature_id))
    r_json = r.json()
    field_photo_path = r_json['fields']['file_list']
    photo_name = field_photo_path.split('/')[-1]
    filepath = photo_directory + '/' + photo_name
    f = open(filepath, 'rb')
    req = requests.put(ngw_api_url + '/component/file_upload/upload', data=f, auth=ngw_auth)
    json_data = req.json()
    json_data['name'] = photo_name
    attach_data = {}
    attach_data['file_upload'] = json_data
    req = requests.post(feature_dst + str(feature_id) + '/attachment/', data=json.dumps(attach_data), auth=ngw_auth)
    print(str(feature_id) + ' status cod: ' + str(req.status_code))
```
